1. Изучить структуру html страницы: https://www.bbcgoodfood.com/recipes/meatball-black-bean-chilli 

2. Придумать структуру в которой хранить данные про рецепт

3. Написать парсер, который:
    * На вход принимает url рецепта
    * На выходе в формате json отдает структурированный рецепт :)

4. Отправка ответа: 
Отправьте свой ответ в форме “Ответов на тестовое задание”.

```python
import requests
from bs4 import BeautifulSoup

def parse_recipe(url, cache=False):
    
    r = requests.get(url)
    if cache:
        with open('cache.html', 'w', encoding='utf-8') as f:
            f.write(r.text)
    
    result = dict()
    soup = BeautifulSoup(r.text)
    
    name_html = (
        soup
        .find('h1', {'class': 'header__title'})
    )
    result['name'] = name_html.text
    
    description_html = (
        soup
        .find('div', {'class': 'header__body'})
        .find('div', {'class': 'editor-content'})
    )
    result['description'] = description_html.text
    
    ingredients_html = (
        soup
        .find('section', {'class': 'recipe__ingredients'})
        .find_all('li')
    )
    result['ingredients'] = [ingredient.text for ingredient in ingredients_html]
    
    steps_html = (
        soup
        .find('section', {'class': 'recipe__method-steps'})
        .find_all('div', {'class': 'editor-content'})
    )
    result['steps'] = [step.text for step in steps_html]
    
    tips_html = (
        soup
        .find('div', {'class': 'post__content-end'})
        .find_all('section', {'class': 'highlight-box'})
    )
    result['tips'] = [tip.text for tip in tips_html]
    
    nutrition_html = (
        soup
        .find('div', {'class': 'header__body'})
        .find_all('tr', {'class': 'key-value-blocks__item'})
    )
    result['nutrition'] = [nutrition.text for nutrition in nutrition_html]
    
    properties_html = (
        soup
        .find('div', {'class': 'header__body'})
        .find_all('div', {'class': 'icon-with-text'})
    )
    result['time'] = properties_html[0].text
    result['level'] = properties_html[1].text
    result['people'] = properties_html[2].text
    
    return result
```

```python
import json

url = 'https://www.bbcgoodfood.com/recipes/meatball-black-bean-chilli'
print(json.dumps(parse_recipe(url), indent=4, sort_keys=True))
```
